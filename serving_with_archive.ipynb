# Problem Statement 

In a single Project import/register multiple model receipes graphs (in .zip format) using V3IO. 

# Import Libraries

```python
import os
import json
import yaml
import mlrun
```

# Create or get project

```python
project = mlrun.get_or_create_project('multiplereceipe-test','./serving/test/',user_project=True)
```

# Testing

```python
serving_test = mlrun.new_function("serving2", kind="serving")
graph = serving_test.set_topology("flow")
graph.to(name="double",handler="double") \
     .to(name="add3", handler="add3") \
     .to(name="echo", handler="echo").respond()

# plot the serving graph
serving_test.spec.graph.plot(rankdir="LR")
```

# Export Graph Yaml for receipe 2

```python
serving_test.export("./serving/test/servingtest.yaml")
```

```python


op2=project.log_artifact(
    "graph_2_definition",
    local_path="./serving/test/servingtest.yaml",
    labels={"id": "receip_test"},
)
print(op2.get_target_path())
```

```python
source2=project.log_artifact(
    "graph_2_source_test",
    local_path="./serving/test/func.zip",
    labels={"id": "receip2"},
)
print(source2.get_target_path())
```

# Import Both Receipe Graphs definitions from V3IO

```python
function2 = mlrun.import_function(url=op2.get_target_path(),project=project)
# print("function1 :::",function1)
# print("-------------------------------------------------------------------------------------------------------------------------------------")
# print("function2 :::",function2)
```

# With Git

```python
from mlrun.platforms import mount_v3io

function2.apply(mlrun.auto_mount())
function2.with_source_archive(source='git://github.com/GiladShapira94/archive.git#master',handler='mylib2:handler',workdir='./func.zip')
function2.spec.build.base_image = 'mlrun/mlrun'

print("function2 :::",function2)
```

```python
function2.deploy()
```

```python
function2.invoke('http://multiplereceipe-test-gilad-serving2-multiplereceipe-test-gilad.default-tenant.app.cust-cs-3-2-3.iguazio-cd2.com/',5)
```

# With V3io

```python
from mlrun.platforms import mount_v3io

function2.apply(mlrun.auto_mount())
function2.with_source_archive(source=source2.get_target_path(),handler='mylib2:handler')
function2.spec.build.base_image = 'mlrun/mlrun'

print("function2 :::",function2)
```

```python
remote=function2.deploy()
```

```python
function2.invoke(remote,5)
```
