# Problem Statement 

In a single Project import/register multiple model receipes graphs (in .zip format) using V3IO. 

# Import Libraries

```python
import os
import json
import yaml
import mlrun
```

# Create or get project

```python
project = mlrun.get_or_create_project('multiplereceipe','./')
```

# Create a serving Graph for model receipe 1

```python
serving = mlrun.code_to_function("serving1", kind="serving", image="mlrun/mlrun")
graph = serving.set_topology("flow")
graph.to(name="double", handler="mylib1.double") \
     .to(name="add3", handler="mylib1.add3") \
     .to(name="echo", handler="mylib1.echo").respond()

# plot the serving graph
serving.spec.graph.plot(rankdir="LR")
```

# Export Graph Yaml for receipe 1

```python
serving.export("serving1.yaml")
```

# Create serving Graph for model receipe 2

```python
serving = mlrun.code_to_function("serving2", kind="serving", image="mlrun/mlrun")
graph = serving.set_topology("flow")
graph.to(name="double", handler="mylib2.double") \
     .to(name="add3", handler="mylib2.add3") \
     .to(name="echo", handler="mylib2.echo").respond()

# plot the serving graph
serving.spec.graph.plot(rankdir="LR")
```

# Export Graph Yaml for receipe 2

```python
serving.export("serving2.yaml")
```

# Save both Receipe Graphs in V3IO

```python
op1=project.log_artifact(
    "graph_1_definition",
    local_path="serving1.yaml",
    labels={"id": "receip1"},
)

op2=project.log_artifact(
    "graph_2_definition",
    local_path="serving2.yaml",
    labels={"id": "receip2"},
)
print(op1.get_target_path())
print(op2.get_target_path())
```

# Import Both Receipe Graphs definitions from V3IO

```python
function1 = mlrun.import_function(url=op1.get_target_path(),project=project)
function2 = mlrun.import_function(url=op2.get_target_path(),project=project)
print("function1 :::",function1)
print("-------------------------------------------------------------------------------------------------------------------------------------")
print("function2 :::",function2)
project.save()
```

# Save Receipe Graph source in V3IO

```python
source1=project.log_artifact(
    "graph_1_source",
    local_path="mylib1.zip",
    labels={"id": "receip1"},
)

source2=project.log_artifact(
    "graph_2_source",
    local_path="mylib2.zip",
    labels={"id": "receip2"},
)
print(source1.get_target_path())
print(source2.get_target_path())
```

# Add source from V3IO to both functions

```python
from mlrun.platforms import mount_v3io

function1.apply(mlrun.auto_mount())
function1.with_source_archive(source=source1.get_target_path(),handler='mylib1.py')

function2.apply(mlrun.auto_mount())
function2.with_source_archive(source=source2.get_target_path(),handler='mylib2.py')

print("function1 :::",function1)
print("-------------------------------------------------------------------------------------------------------------------------------------")
print("function2 :::",function2)
```
